<h4> 1.4 Calculating mean median mode and Standard Deviation from DataFrames </h4><br>
<p> Lets compute mean, median,mode and Standard Deviation from Data Frames </p>
<img src='https://www.careerpower.in/images/mean-median-mode-formula.png'>

```python
import random

def generate_data_frame():
    df=pd.DataFrame(columns=['X','Y','Z'])
    n=int(input('Enter number of rows: '))
    for cord in df.columns:
        x=[random.randrange(0,100) for i in range(0,n)]
        df[cord]=x
    return df 

def get_m_m_m_SD(df,f):
    if f==1:
        return df['X'].mean(),df['Y'].mean(),df['Z'].mean()
    elif f==2:
        return df['X'].median(),df['Y'].median(),df['Z'].median()
    elif f==4:
        return df['X'].std(ddof=0),df['Y'].std(ddof=0),df['Z'].std(ddof=0)
    else: 
        return df['X'].mode(),df['Y'].mode(),df['Z'].mode()
    
df=generate_data_frame()
df
```

<h4> 2.2. Bargraphs </h4>
<br>
<p> Understanding how to build Bargraphs from 2D data </p>

```python
def load_data():
    df=pd.DataFrame()
    df['S.N']=[1,2,3,4,5]
    df['Name']=['Siddhant','Prashant','Prakriti','Pukar','Aakash']
    df['Maths']=[70,80,85,72,65]
    df['Science']=[65,55,75,70,82]
    df['Literature']=[5,72,30,45,50]
    return df
```

```python
df=load_data()
fig=plt.figure(figsize=(15,15))

#plot 1: 
plt.subplot(221) #row column index
plt.bar(df['Name'],df['Maths'])
plt.xlabel('Students')
plt.ylabel('Marks_Maths')

#plot 2: 
plt.subplot(222) #row column index
plt.bar(df['Name'],df['Science'])
plt.xlabel('Students')
plt.ylabel('Marks_Science')

#plot 3: 
plt.subplot(223) #row column index
plt.bar(df['Name'],df['Literature'])
plt.xlabel('Students')
plt.ylabel('Marks_Literature')


fig.tight_layout(pad=1.0)
```

<h3> 3. Combining 1 and 2 </h3> <br> 
<p> Lets combine the Matplotlib basics with pandas basics to have dataframe visualize it and also get key insights from it </p>

```python
df=generate_data_frame()
X=df['X']
Y=df['Y']
Z=df['Z']
x_mean,y_mean,z_mean=get_m_m_m_SD(df,1)

fig=plt.figure(figsize=(15,15))

plt.subplot(321)
plt.title('X-Y with vline mean point')
plt.scatter(X,Y)
plt.scatter(x_mean,y_mean,color='red')
plt.axvline(x=x_mean)

plt.subplot(322)
plt.title('X-Y with hline mean point')
plt.scatter(X,Y)
plt.scatter(x_mean,y_mean,color='red')
plt.axhline(y=y_mean)

plt.subplot(323)
plt.title('Y-Z with vline mean point')
plt.scatter(Y,Z)
plt.scatter(y_mean,z_mean,color='red')
plt.axvline(x=y_mean)

plt.subplot(324)
plt.title('Y-Z with vline mean point')
plt.scatter(Y,Z)
plt.scatter(y_mean,z_mean,color='red')
plt.axhline(y=z_mean)

plt.subplot(325)
plt.title('X-Z with vline mean point')
plt.scatter(X,Z)
plt.scatter(x_mean,z_mean,color='red')
plt.axvline(x=x_mean)

plt.subplot(326)
plt.title('X-Z with vline mean point')
plt.scatter(X,Z)
plt.scatter(x_mean,z_mean,color='red')
plt.axhline(y=z_mean)
```
